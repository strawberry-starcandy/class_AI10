```python
# 단순 링크드 리스트 
# 학생들의 이름과 성적을 입력 받아 성적순으로 클래스의 data에 딕셔너리로 저장
class Student:
    def __init__(self):
        self.name = None
        self.score = None
        self.link = None
# 성적순으로 링크드 리스트 작성
# insert_node() 수정
# 1. 학생 정보 입력
# 2. 학생 정보 삭제 ( 이름을 입력 받음 )
# 3. 학생 정보 검색 ( 이름 입력 받음 )
# 4. 학생 전체 정보 출력 ( 등수도 출력 )
# 9. 프로그램 종료
```

```python
def menu_print():
    print('==================================')
    print('        학생 관리 프로그램')
    print(' 1. 학생 정보 입력 ')
    print(' 2. 학생 정보 삭제 ')
    print(' 3. 학생 정보 검색 ')
    print(' 4. 학생 정보 조회 ')
    print(' 9. 프로그램 종료 ')  
    print('==================================')    
    print('선택 > ',end='')
    menu = input()
    return menu

def add_node(node):
    global head, tail
    if head == None: 
        head = node
        tail = node     
        return
     
    curr_node = head
    prev_node = curr_node
    while curr_node.link:
        if curr_node.score > node.score:
            if curr_node == head:   # 처음 자료 앞에 삽입
                node.link = head
                head = node
            else:
                node.link = curr_node   # 중간 노드 삽입
                prev_node.link = node
            return
        
        prev_node = curr_node
        curr_node = curr_node.link
        
    # 마지막 노드의 값과 비교 후 삽입
    if curr_node.score > node.score:
        node.link = curr_node   # 마지막 노드 앞 삽입
        prev_node.link = node
    else:
        curr_node.link = node
        tail = node
    return

def del_node(name):
    global head, tail
    node = head
    prev_node = node
    while node:
        if node.name == name:
            if node == head:
                head = node.link
            elif node == tail:
                prev_node.link = None
                tail = prev_node
            else:
                prev_node.link = node.link
            del node
            return
            
        prev_node = node
        node = node.link
            
def list_node():
    node = head
    while node:        
        print(f'{node.name} : {node.score}')
        node = node.link
```
