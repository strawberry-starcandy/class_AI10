```python
# 원형 링크드 리스트 작성
# 처음 node 생성시 -> head 에 node 저장. node.link에 node 저장
# 마지막에 node 삽입시 -> 마지막노드 link에 node 저장
#                         head값을 node.link에 저장
# 중간 node 삽입시 -> node.link에 현재노드 저장
#                     이전노드의 link에 node 저장, 
# 처음 node 삽입시 -> head에서 시작해서 마지막 노드 검색 
#                     마지막 node.link 에 node 저장
#                    node.link 에 head 저장, head에 node 저장
```

```python
class Node:
    seq = 0
    def __init__(self):
        self.no = None
        self.data = None
        self.link = None 
```

```python
def menu_prt():
    print(' 1. 자료입력 -> ')
    print(' 2. 자료삭제 ')
    print(' 3. 자료 조회  ')  
    print(' 9. 프로그램 종료 ')  
    print('==================================')    
    print('선택 > ',end='')
    menu = input()
    return menu

def add_node():
    global HEAD
    # node 생성
    node = Node()     # 빈 node 생성
    node.data = input('이름 입력 : ')  # node.data 에 자료 입력
    Node.seq += 1
    node.no = Node.seq
    # 처음 node 생성시 -> head 에 node 저장. node.link에 node 저장
    if HEAD == None:
        HEAD = node
        node.link = node
        return 
    elif HEAD.data > node.data: # HEAD 앞에 삽입 
        last = HEAD
        while last.link != HEAD:   # 마지막 node 검색
            last = last.link
        last.link = node
        node.link = HEAD
        HEAD = node
        return 
    
    curr = HEAD  
    while curr.link != HEAD:
        if curr.data > node.data:             # 중간 삽입
            node.link = curr     # 삽입 node.link 에 현재노드 저장
            prev.link = node     # 이전 node.link 에 node 저장
            return
        prev = curr        # 현재 node를 prev로 저장 
        curr = curr.link   # 다음 node로 이동 

   # 마지막 노드 삽입 마지막노드 link에 node 저장,head값을 node.link에 저장 
    if  curr.data > node.data:
        # 중간 삽입
        node.link = curr     # 삽입 node.link 에 현재노드 저장
        prev.link = node  
    else:
        curr.link = node
        node.link = HEAD
              
def del_node():
    global HEAD
    name = input('삭제할 자료 입력 -> ')
    # 처음 node 삭제 -> HEAD = HEAD.link
    # 마지막 node.link 에 HEAD.link 
    if HEAD.data == name:
        last = HEAD
        while last.link != HEAD:  # 마지막 node 검색
            last = last.link    
        HEAD = HEAD.link
        del last.link  # 메모리 삭제
        last.link = HEAD         
        return
    
    curr = HEAD
    while curr.link != HEAD:
        # 중간 node 삭제 -> prev.link = curr.link
        if curr.data == name:
            prev.link = curr.link
            del curr     # curr node 메모리 삭제
            return       
        prev = curr
        curr = curr.link
        
   # 마지막 node 삭제  -> prev.link = curr.link
    prev.link = curr.link
    del curr    # curr node 메모리 삭제

def list_node():
    curr = HEAD
    if HEAD == None:
        return
    while curr.link != HEAD:
        print(f'{curr.no} -> {curr.data}')
        curr = curr.link
    print(f'{curr.no} -> {curr.data}')
```

```python
HEAD = None
while True:
    menu = menu_prt()
    if menu == '1':
        add_node()
    elif menu == '2':   
        del_node()
    elif menu == '3':
        list_node()
    elif menu == '9':
        break
```
