```python
def QuickSort_ex(arr, start, end):
    if start >= end:
        return arr

    low = start
    high = end

    pivot = arr[(low + high) // 2]
    while low < high:
        while arr[low] < pivot:
            low += 1
        while arr[high] > pivot:
            high -= 1
        if low <= high:
            arr[low], arr[high] = arr[high], arr[low]
            low += 1
            high -= 1
    mid = low
    QuickSort_ex(arr, start, mid - 1)
    QuickSort_ex(arr, mid, end)
```

```python
import random
array = [ random.randrange(1,1000000) for _ in range(10000) ]

print('quick sort before --> ', array)
QuickSort_ex(array, 0, len(array) - 1)
print('quick sort after --> ', array)
```
