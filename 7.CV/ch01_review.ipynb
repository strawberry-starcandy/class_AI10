```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import sys
import os
```

```python
# img = cv2.imread('./Ch_1_fig/fig/puppy.bmp', cv2.IMREAD_COLOR)
img = cv2.imread('./Ch_1_fig/fig/puppy.bmp', cv2.IMREAD_GRAYSCALE)

if img is None:
    print('image read failed')
    exit()

img_resize = cv2.resize(img, (2000, 1400), interpolation=cv2.INTER_LINEAR)
cv2.imwrite('./Ch_1_fig/fig/img_zoom.png', img_resize)

cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image', img)
cv2.imshow('image_resize', img_resize)

while True:
    key = cv2.waitKey()
    if key == 27 or key == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
imgBGR = cv2.imread('./Ch_1_fig/fig/puppy.bmp')
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)

fig, ax = plt.subplots(1,2)
img_list = [imgBGR, imgRGB]

for i, j in enumerate(img_list):
    ax[i].imshow(j)
    ax[i].axis('off')

plt.show()
```

```python
img_lists = os.listdir('./Ch_1_fig/fig/images/')
# print(img_lists)

img_files = []
for i in img_lists:
    img_path = './Ch_1_fig/fig/images/' + i
    img_files.append(img_path)

print(img_files)

cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

idx = 0
while True:
    img = cv2.imread(img_files[idx])

    if img is None:
        print('image read failed')
        break
    cv2.imshow('scene', img)

    if cv2.waitKey(3000) == 27:
        break
    
    idx = (idx + 1) % len(img_files)
cv2.destroyAllWindows()
```
