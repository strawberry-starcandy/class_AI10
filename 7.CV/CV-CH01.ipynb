```python
img = cv2.imread('./Ch_1_fig/fig/puppy.bmp', cv2.IMREAD_COLOR)
print(type(img))
print(img.shape)

if img is not None:
    cv2.namedWindow('image')
    cv2.imshow('image', img)
    key = cv2.waitKey()
    
    while True:
        if 49 <= key <= 57:
            num = key - 48
            cv2.imshow('image', cv2.resize(img, (img.shape[1] * num, img.shape[0] * num)))
            key = cv2.waitKey()
            # print(key, type(key))
        else:
            break

    cv2.destroyAllWindows()
```

```python
img = cv2.imread('./Ch_1_fig/fig/puppy.bmp', cv2.IMREAD_COLOR)

if img is None:
    print('Image read failed')
    sys.exit()
print(type(img))
print(img.shape)

cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE)
cv2.imshow('image', img)
cv2.waitKey(2000)
cv2.destroyWindow('image')

img_resize = cv2.resize(img, (320, 240), interpolation=cv2.INTER_LINEAR)
print(type(img_resize))
print(img_resize.shape)

cv2.namedWindow('image_resized', cv2.WINDOW_AUTOSIZE)
cv2.imshow('image_resized', img_resize)
k = cv2.waitKey(2000)
print(k)

cv2.destroyAllWindows()

cv2.imwrite('./Ch_1_fig/fig/puppy_resized.png', img_resize)
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Ch_1_fig/fig/puppy.bmp', cv2.IMREAD_COLOR)

if img is None:
    print('image read failed')
    sys.exit()

img_plt = plt.imread('Ch_1_fig/fig/puppy.bmp')
plt.figure(figsize=(4,4))
plt.axis('off')
plt.title('CV Image')
plt.imshow(img)
plt.show()

plt.figure(figsize=(4,4))
plt.axis('off')
plt.title('CV Image(Indexing)')
plt.imshow(img[:,:,::-1])
plt.show()

imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(4,4))
plt.axis('off')
plt.title('CV Image(cvtColor)')
plt.imshow(imgRGB)
plt.show()

plt.figure(figsize=(4,4))
plt.axis('off')
plt.title('CV Image(plt read)')
plt.imshow(img_plt)
plt.show()

cv2.namedWindow('image')
cv2.imshow('image', img)

while True:
    k = cv2.waitKey()
    if k == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
img_list_pre = os.listdir('./Ch_1_fig/fig/images/')
# print(img_list_pre)

img_lists = []
for i in img_list_pre:
    img_path = './Ch_1_fig/fig/images/' + i
    img_lists.append(img_path)

# print(img_lists)
cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

k = -1
while k != 27:
    for i in img_lists:
        img = cv2.imread(i)
        
        if img is None:
            print(f'{i} read failed')
            break
        
        cv2.imshow('scene', img)

        k = cv2.waitKey(500)
        if k == 27:
            break
cv2.destroyAllWindows()
```

```python
tmp = []
tmp.extend(np.arange(0.1, 1.1,0.05))
tmp.extend(np.ones(20))
tmp.extend(np.arange(1, 0,-0.1))
print(tmp)

img_list_pre = os.listdir('./Ch_1_fig/fig/my_images/')
# print(img_list_pre)

img_lists = []
for i in img_list_pre:
    img_path = './Ch_1_fig/fig/my_images/' + i
    img_lists.append(img_path)

# print(img_lists)
cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

k = -1
while k != 27:
    for i in img_lists:
        img = cv2.imread(i)
        
        if img is None:
            print(f'{i} read failed')
            continue

        for t in tmp:
            img_tmp = img / 255 * t
            cv2.imshow('scene', img_tmp)

            k = cv2.waitKey(20)
            if k == 27:
                break
            if k == 32:
                break
        if k == 27:
            break
cv2.destroyAllWindows()
```
