```python
## 카메라 동작 반전 플립 등등
cap = cv2.VideoCapture(0)

key_cmd = {
    ord('e'):lambda cur, src, a:cv2.Canny(cur, 50, 150),
    ord('h'):lambda cur, src, a:cur[::-1],
    ord('v'):lambda cur, src, a:cur[:,::-1],
    ord('i'):lambda cur, src, a:255 - cur
}

cur_flag = {
    ord('e'):0,
    ord('h'):0,
    ord('v'):0,
    ord('i'):0
}

bias = 0

if not cap.isOpened():
    print('Video capture failed')
    sys.exit()

w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

fps = int(cap.get(cv2.CAP_PROP_FPS))

# fourcc = cv2.VideoWriter_fourcc(*'DIVX')
# out = cv2.VideoWriter('cur_frame.avi', fourcc, fps, (w, h))

while True:
    ret, frame = cap.read()
    if not ret:
        print('frame read failed')
        break
    
    cur_frame = frame.copy()
    
    for k, flag in cur_flag.items():
        if flag:
            cur_frame = key_cmd[k](cur_frame, cur_frame, 0)
    
    if bias != 0:
        if len(cur_frame.shape) ==2:
            cur_frame = cv2.add(cur_frame, bias, )
        else:
            cur_frame = cv2.add(cur_frame, (bias, bias, bias, bias), )
    
    # out.write(cur_frame)
    cv2.imshow('Cam00', cur_frame)
    key = cv2.waitKey(1000 // fps)
    
    if key == 27:
        break
    elif key == ord(' '):
        bias = 0
        for k in cur_flag.keys():
            cur_flag[k] = False
    elif key in key_cmd.keys():
        cur_flag[key] = 1 if cur_flag[key] == 0 else 0
        # print(cur_flag)
    elif key in [ord('+'), ord('-')]:
        op =  1 if key == ord('+') else -1
        bias += 10 * op
        if abs(bias) > 100:
            bias -= (abs(bias) - 100) * op
        # print(bias, op)

# out.release()
cap.release()
cv2.destroyAllWindows()
```
