```python
html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>
"""
```

```python
from bs4 import BeautifulSoup
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20221208'
soup = BeautifulSoup(request.urlopen(url))
print(soup.prettify())
```

```python
import requests
resp = requests.get(url)
soup = bs(resp.text)
print(soup.prettify())
```

```python
movie_titles = []
movie_points = []
for line in soup.find_all('tr'):
    movie_title = line.find('div', class_='tit5')
    #print(movie_title)
    if movie_title:
        movie_titles.append(movie_title.get_text().strip())
        
    movie_point = line.find('td', class_='point')
    if movie_point:
        movie_points.append(movie_point.get_text().strip())
print(movie_titles) 
print(movie_points)
print()
for idx, (title, point) in enumerate(zip(movie_titles, movie_points)):
    print(f'{idx + 1:3d} : {title:100s}\t - {point}')
```

```python
## 다음 뉴스에서 20221209113834465, 20221209124242834 해당 기사의 제목과 내용 출력

def news_get(topic):
    contents = []
    url = 'https://v.daum.net/v/'
    resp = requests.get(url + topic)
    soup = bs(resp.text)
    
    news_title = soup.find('h3', class_ = 'tit_view')
    if news_title:
        contents.append(news_title.get_text().strip())
#     news_summary = soup.find('strong', class_ = 'summary_view')
#     if news_summary:
#         contents.append(news_summary.get_text().strip())

    article_body = soup.find('div', class_ = 'news_view fs_type1')
    for line in article_body.find_all('p'):
        if line:
            contents.append(line.get_text().strip())
    
    return contents

test = news_get('20221209113834465')
print(*test, sep='\n')
print()
print('-' * 128)
print()

test = news_get('20221209124242834')
print(*test, sep='\n')
```
