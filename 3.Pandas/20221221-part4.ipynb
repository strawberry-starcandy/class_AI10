```python
# %load matplot.py
import matplotlib.pyplot as plt
# matplotlib 폰트설정
plt.rc('font', family='D2Coding')

```

```python
import pandas as pd

df = pd.read_excel('./data/시도별 전출입 인구수.xlsx')
df
```

```python
# 1. 누락 데이터 처리 : df.fillna(value or method=ffill/bfill)
df = df.fillna(method='ffill')
print(df.head())
```

```python
# 서울에서 다른 지역으로 전출한 인구수의 분포도
# 2. 전출지별이 서울특별시이고, 전입지별이 서울특별시가 아닌 자료만 추출
```

```python
df_seoul = df[(df.전출지별 == '서울특별시') & (df.전입지별 != '서울특별시')]
df_seoul = df_seoul.replace('-',0)
```

```python
# 컬럼 또는 인덱스 삭제 -> df.drop(컬럼명, axis=1), df.drop(인덱스명, axis=0)
df_seoul = df_seoul.drop('전출지별', axis=1)
df_seoul.head()
```

```python
# 4. 컬럼명 재정의 : '전입지별' -> '전입지'로 변경
# df_seoul.rename(columns={'전입지별':'전입지'}, inplace=True)
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul.head()
```

```python
df_seoul.set_index('전입지', inplace=True)
```

```python
# 최종 : 서울에서 경기도로 전입한 인구의 이동 데이터 추출
df_one = df_seoul.loc['경기도']
df_one
```

```python
plt.plot(df_one.index, df_one.values)
plt.title('서울->경기도 인구 이동')
plt.xlabel('기간')
plt.ylabel('이동 인구수')
plt.show()
```

```python
# 남북한발전전략.xlsx 파일을 읽어서 df 저장 후
# Nan -> ffill로 대체
# 남한과 북한의 합계 자료만 추출
# '전력량 (억kWh)' -> '구분' 컬럼명으로, '발전 전력별'은 삭제
# 구분을 인덱스로
# 그래프 그리기 ->
#    제 목: 남한과 북한의 연도별 전력 사용량
#    xlabel: 기간
#    ylabel: 사용량
```

```python
df = pd.read_excel('./data/남북한발전전력량.xlsx').fillna(method='ffill')
df_ns = df[df['발전 전력별'] == '합계']
df_ns_tot = df_ns.rename({df_ns.columns[0]:'구분'}, axis=1).drop(df_ns.columns[1], axis=1)
df_graph = df_ns_tot.set_index(df_ns_tot.columns[0]).T
df_graph.남한.map(int);
df_graph.북한.map(int);
```

```python
# 히스토그램 / 커널 밀도 그래프 - 하나의 변수 데이터의 분포를 확인
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

# distplot()
sns.distplot(titanic['fare'], ax=ax1)

# kdeplot()
sns.kdeplot(x='fare', data=titanic, ax=ax2)

# histplot()
sns.histplot(x='fare', data=titanic, ax=ax3)

ax1.set_title('titanic fare -- hist/kde')
plt.show()
```

```python
# heatmap -> 데이터프레임을 피벗테이블로 정리하고 할 때
table = titanic.pivot_table(index=['sex'], columns=['class'], aggfunc='size')
```

```python
sns.heatmap(table, annot=True, fmt='d', cmap = 'YlGnBu', linewidths=1, cbar=True)
plt.show()
```

```python
# 범주형 데이터의 산점도 -> stripplot(), swarmplot(): 데이터의 분산 정도도 표현
sns.set_style('whitegrid')
fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.stripplot(x='class',y='age',data=titanic, ax=ax1)
sns.swarmplot(x='class',y='age',data=titanic, ax=ax2)
plt.show()
```

```python
# 막대 그래프 -> barplot
fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

# x축 y축 설정
sns.barplot(x='sex', y='survived', data=titanic, ax=ax1)
# hue 변수 설정
sns.barplot(x='sex', y='survived', data=titanic, ax=ax2, hue='class')
sns.barplot(x='sex', y='survived', data=titanic, ax=ax3, hue='class', dodge=False)
plt.show()
```

```python
# 빈도 그래프 -> countplot()
fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

# 기본값
sns.countplot(x='class', palette='Set1', data=titanic, ax=ax1)

# hue 옵션에 'who' 추가
sns.countplot(x='class', hue='who', palette='Set2', data=titanic, ax=ax2)

#dodge=False 옵션 추가(축 방향으로 분리하지 않고 누적 그래프 출력)
sns.countplot(x='class', hue='who', palette='Set3', dodge=False, data=titanic, ax=ax3)

# 차트 제목 표시
ax1.set_title('titanic class')
ax2.set_title('titanic class - who')
ax3.set_title('titanic class - who(stacked)')

plt.show()
```
