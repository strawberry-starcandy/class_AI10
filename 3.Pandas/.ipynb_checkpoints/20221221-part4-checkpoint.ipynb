```python
# %load matplot.py
import matplotlib.pyplot as plt
# matplotlib 폰트설정
plt.rc('font', family='D2Coding')

```

```python
import pandas as pd

df = pd.read_excel('./data/시도별 전출입 인구수.xlsx')
df
```

```python
# 1. 누락 데이터 처리 : df.fillna(value or method=ffill/bfill)
df = df.fillna(method='ffill')
print(df.head())
```

```python
# 서울에서 다른 지역으로 전출한 인구수의 분포도
# 2. 전출지별이 서울특별시이고, 전입지별이 서울특별시가 아닌 자료만 추출
```

```python
df_seoul = df[(df.전출지별 == '서울특별시') & (df.전입지별 != '서울특별시')]
df_seoul = df_seoul.replace('-',0)
```

```python
# 컬럼 또는 인덱스 삭제 -> df.drop(컬럼명, axis=1), df.drop(인덱스명, axis=0)
df_seoul = df_seoul.drop('전출지별', axis=1)
df_seoul.head()
```

```python
# 4. 컬럼명 재정의 : '전입지별' -> '전입지'로 변경
# df_seoul.rename(columns={'전입지별':'전입지'}, inplace=True)
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul.head()
```

```python
df_seoul.set_index('전입지', inplace=True)
```

```python
# 최종 : 서울에서 경기도로 전입한 인구의 이동 데이터 추출
df_one = df_seoul.loc['경기도']
df_one
```

```python
plt.plot(df_one.index, df_one.values)
plt.title('서울->경기도 인구 이동')
plt.xlabel('기간')
plt.ylabel('이동 인구수')
plt.show()
```

```python
# 남북한발전전략.xlsx 파일을 읽어서 df 저장 후
# Nan -> ffill로 대체
# 남한과 북한의 합계 자료만 추출
# '전력량 (억kWh)' -> '구분' 컬럼명으로, '발전 전력별'은 삭제
# 구분을 인덱스로
# 그래프 그리기 ->
#    제 목: 남한과 북한의 연도별 전력 사용량
#    xlabel: 기간
#    ylabel: 사용량
```

```python
df = pd.read_excel('./data/남북한발전전력량.xlsx').fillna(method='ffill')
df_ns = df[df['발전 전력별'] == '합계']
df_ns_tot = df_ns.rename({df_ns.columns[0]:'구분'}, axis=1).drop(df_ns.columns[1], axis=1)
df_graph = df_ns_tot.set_index(df_ns_tot.columns[0]).T
df_graph.남한.map(int);
df_graph.북한.map(int);
```

```python
plt.figure(figsize=(14,5))
plt.plot(df_graph.index, df_graph.values, '-*')
plt.title('남한과 북한의 연도별 전력 사용량', size=15)
plt.xlabel('기간')
plt.ylabel('사용량')
plt.xticks(rotation='vertical')
plt.legend(labels=df_graph.columns)
plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.plot(df_one.index, df_one.values, marker='o', markersize=5, c='r')
plt.title('서울->경기도 인구 이동', size=15)
plt.xlabel('기간', size=10)
plt.ylabel('이동 인구수', size=12)
plt.legend(labels=['서울 -> 경기'])
plt.xticks(rotation=60)
# y축 범위 지정 (최소값, 최대값)
plt.ylim(50000, 800000)

# 주석 표시 - 화살표
plt.annotate('',
             xy=(20, 600000),       #화살표의 머리 부분(끝점)
             xytext=(2, 270000),    #화살표의 꼬리 부분(시작점)
             xycoords='data',       #좌표체계
             arrowprops=dict(arrowstyle='->', color='skyblue', lw=5), #화살표 서식
             )

plt.annotate('',
             xy=(47, 450000),       #화살표의 머리 부분(끝점)
             xytext=(30, 580000),   #화살표의 꼬리 부분(시작점)
             xycoords='data',       #좌표체계
             arrowprops=dict(arrowstyle='->', color='olive', lw=5),  #화살표 서식
             )

# 주석 표시 - 텍스트
plt.annotate('인구이동 증가(1970-1995)',  #텍스트 입력
             xy=(10, 360000),            #텍스트 위치 기준점
             rotation=25,                #텍스트 회전각도
             va='baseline',              #텍스트 상하 정렬
             ha='center',                #텍스트 좌우 정렬
             fontsize=15,                #텍스트 크기
             )

plt.annotate('인구이동 감소(1995-2017)',  #텍스트 입력
             xy=(39, 510000),            #텍스트 위치 기준점
             rotation=-11,               #텍스트 회전각도
             va='baseline',              #텍스트 상하 정렬
             ha='center',                #텍스트 좌우 정렬
             fontsize=15,                #텍스트 크기
            )
plt.show()
```

```python
# 서울에서 경기도 이동 인구수를 다른 형식으로 2개를 그림
fig = plt.figure(figsize=(12,12))

# 그래프 추가 -> fig.add_subplot(행의 그래프 개수, 열의 그래프 개수, 순번)
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

# ax1 에 그래프 그리기
ax1.plot(df_one, marker='o')
ax2.plot(df_one, marker='*', markerfacecolor='green', color='olive', label='서울 -> 경기')
ax2.legend(loc='best')

# y축 범위 지정
ax1.set_ylim(50000, 800000)
ax2.set_ylim(50000, 800000)

# x축 눈금 라벨 지정 및 회전(90)
ax1.set_xticklabels(df_one.index, rotation=90)
ax2.set_xticklabels(df_one.index, rotation=70)
plt.show()
```

```python
# 서울에서 충청남도, 경상북도, 강원도로 각각 이동한 인구수를 그래프로 표현
# 하나에 3개의 그래프로 표현
df_seoul_g = df_seoul.loc[['충청남도', '경상북도', '강원도']]
```

```python
fig = plt.figure(figsize=(12,15))

# 그래프 추가 -> fig.add_subplot(행의 그래프 개수, 열의 그래프 개수, 순번)
ax1 = fig.add_subplot(3, 1, 1)
ax2 = fig.add_subplot(3, 1, 2)
ax3 = fig.add_subplot(3, 1, 3)

# 그래프 그리기
ax1.plot(df_seoul_g.loc['충청남도'], marker='o',color='r')
ax2.plot(df_seoul_g.loc['경상북도'], marker='*', color='g')
ax3.plot(df_seoul_g.loc['강원도'], marker='^', color='b')

# y축 범위 지정
ax1.set_ylim(5000, 60000)
ax2.set_ylim(5000, 60000)
ax3.set_ylim(5000, 60000)

# x축 눈금 라벨 지정 및 회전
ax1.tick_params(axis='x', rotation=90)
ax2.tick_params(axis='x', rotation=90)
ax3.tick_params(axis='x', rotation=90)

plt.show()
```

```python
# 하나의 figure, 하나의 axes에 3개의 그래프를 그림
ax = plt.plot(df_seoul_g.T)
plt.legend(df_seoul_g.index)
plt.xticks(rotation=90);
```

```python
# 서울 -> 충남, 충북, 전북, 전남으로 이동한 인구수를 각각의 그래프로 표현
# 연도는 1990년도부터 2017년까지 -> int형으로 변경
# 각각 title, legend, xtick은 75도 회전
# xlabel은 기간, ylabel은 이동 인구수
```

```python
df_seoul_g = df_seoul.loc[['충청남도', '충청북도', '전라북도', '전라남도']]
df_seoul_g.columns = map(int, df_seoul_g.columns)
```

```python
df_seoul_T = df_seoul_g.T
df_4 = df_seoul_T.loc[list(range(1990, 2018))]
df_4T = df_4.T
```

```python
fig = plt.figure(figsize=(12, 5))

ax = [fig.add_subplot(2, 2, x) for x in range(1, len(df_4T.index) + 1)]

markers = ['o', '*', '^', 's']
colors = ['r', 'g', 'b', 'm']
titles = ['충남', '충북', '전북', '전남']
for i in range(len(df_4T.index)):
    ax[i].plot(df_4T.loc[df_4T.index[i]], marker=markers[i],color=colors[i])
    ax[i].set_title('서울 -> ' + titles[i])
```

```python
df_4.plot(kind='area', figsize=(10,5), alpha=0.5, stacked=False)
plt.title('서울 -> 타시도 인구 이동')
plt.ylabel('이동 인구')
plt.xlabel('기간')
plt.tick_params(axis='x', rotation=75)
plt.legend()
```
