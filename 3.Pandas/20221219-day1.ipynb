```python
import pandas as pd
```

```python
# 4 산술연산
# 4-1 시리즈 연산
# 시리즈 연산자 숫자 : 시리즈 + - * / 숫자 -> 시리즈의 각 요소에 연산자 적용
student1 = pd.Series({'국어':100, '영어':90, '수학':80})
print(student1)

print(student1 + 100)
print(student1 / 200)
```

```python
# 시리즈 연산자(+, -, *, /) 시리즈
student2 = pd.Series({'수학':80, '국어':90, '영어':80})
print(student1)    # 국어, 영어, 수학
print(student2)    # 수학, 국어, 영어

# 학생의 과목별 합계를 구함
hap = student1 + student2
cha = student1 - student2
mul = student1 * student2
div = student1 / student2
print(hap)
```

```python
result = pd.DataFrame([hap, cha, mul, div], index=['+', '-', '*', '/'])
result
```

```python
import numpy as np
student1 = pd.Series({'국어':np.nan, '영어':80, '수학':90})
student2 = pd.Series({'수학':90, '국어':80})
print(student1)
print(student2)
```

```python
# 학생의 과목별 합계를 구함
hap = student1 + student2
cha = student1 - student2
mul = student1 * student2
div = student1 / student2

# 연산식에 NaN 이 포함되면 결과는 NaN
result = pd.DataFrame([hap, cha, mul, div], index=['+', '-', '*', '/'])
result
```

```python
# 연산 메소드 -> sr1.add(s2)
hap = student1.add(student2, fill_value=0)    # 연산 메소드 ( fill_value=값) - > NaN을 대체할 값
cha = student1.sub(student2, fill_value=0)
mul = student1.mul(student2, fill_value=0)
div = student1.div(student2, fill_value=0)

result = pd.DataFrame([hap, cha, mul, div], index=['+', '-', '*', '/'])
result
```

```python
# 4-2 데이터프레임 연산
import seaborn as sns    # seaborn에서 제공하는 dataset 을 활용하기 위해 import

# 데이터 프레임 연산자 (+, -, *, /) 숫자
# titanic data 를 load
titanic = sns.load_dataset('titanic')

# 'age', 'fare' 컬럼의 정보만 추출해서 가져옴
df = titanic.loc[:, ['age', 'fare']]

df.head()      # 처음 5행만 추출
df.head(10)    # 처음 10행만 추출

df.tail()      # 마지막 5행만 추출
df.tail(10)    # 마지막 10행만 추출
```

```python
# 데이터프레임 연산자(+, -, *, /) 데이터프레임
result = df_add10 - df
result
```
