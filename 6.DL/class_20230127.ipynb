```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 0.05)
y = np.log(x)
y1 = -np.log(x)
y2 = -np.log(-x)
y3 = -np.log(1-x)

plt.plot(x, y, label='log(x)')
plt.plot(x, y1, label='-log(x)')
plt.plot(x, y2, label='-log(-x)')
plt.plot(x, y3, label='-log(1-x)')

plt.xlim((-2,2))
plt.ylim((0-2,2))

plt.legend()
plt.grid(True)
plt.show()
```

```python
list_y = [y, y1, y2, y3]

fig, ax = plt.subplots(2,2)

ax[0,0].plot(x,y)
ax[0,1].plot(x,y1)
ax[1,0].plot(x,y2)
ax[1,1].plot(x,y3)

ax[0,0].set_xlim(-2,2)
ax[0,0].set_ylim(-2,2)
ax[0,1].set_xlim(-2,2)
ax[0,1].set_ylim(-2,2)
ax[1,0].set_xlim(-2,2)
ax[1,0].set_ylim(-2,2)
ax[1,1].set_xlim(-2,2)
ax[1,1].set_ylim(-2,2)

ax[0,0].grid('on')
ax[0,1].grid('on')
ax[1,0].grid('on')
ax[1,1].grid('on')

ax[0,0].set_title('ln(x)')
ax[0,1].set_title('-ln(x)')
ax[1,0].set_title('-ln(-x)')
ax[1,1].set_title('-ln(1-x)')
plt.show()
```

```python
import pandas as pd
```

```python
speed = [2, 4, 5, 8, 6]
dist = [12, 16, 10, 20, 22]

df = pd.DataFrame(data=[speed, dist]).T
df.columns = ['speed', 'dist']
speed_mean = df.speed.mean()
dist_mean = df.dist.mean()
df
```

```python

divisor = ((df.speed - speed_mean)**2).sum()
divided = ((df.speed - speed_mean) * (df.dist - dist_mean)).sum()

theta1 = divided / divisor
theta0 = dist_mean - theta1 * speed_mean
print(divisor, divided)
print(theta1, theta0)
```

```python
x_hat = np.arange(2, 9)
y_hat = x_hat * theta1 + theta0

plt.scatter(df.speed, df.dist)
plt.plot(x_hat, y_hat, c='r')
plt.show()
```

```python
x = np.arange(2, 10, step=2)
y = np.array([81, 93, 91, 97])
print(x, y)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
mx = np.mean(x)
my = np.mean(y)
print(mx,my)
```

```python
for i in x:
    print(f'i:{i}, {(i - mx)**2}')
```

```python
## Sxx, Syy
Sxx = sum([(i - mx) ** 2 for i in x])
Sxy = sum([(i - mx) * (j - my) for i,j in zip(x,y)])

print(Sxx, Sxy)
```

```python
slope = Sxy / Sxx  # theta 1
intercept = my - (mx * slope) # theta 0

print(f'y = {slope}x + {intercept}')
```

```python
x_ = np.arange(2, 9)
y_hat = slope * x_ + intercept

plt.scatter(x, y,s=80)
plt.plot(x_, y_hat, c='r')
plt.show()
```

```python
mse = 0
for i, j in zip(x, y):
    predicted = slope * i + intercept
    mse += (j - predicted) ** 2
print(round(mse/len(x), 2))
```
